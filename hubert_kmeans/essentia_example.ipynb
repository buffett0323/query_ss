```python
import essentia
import essentia.standard as es
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = '/mnt/gestalt/database/beatport/audio/audio/electro-house/ff527b1b-305f-4893-b8aa-0dd84b03c2ac.mp3'
loader = essentia.standard.MonoLoader(filename=file_path)

# and then we actually perform the loading:
audio = loader()
print(audio)
print(len(audio))
```

```python
import IPython
IPython.display.Audio(file_path)
```

```python
from pylab import plot, show, figure, imshow
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 6) # set plot sizes to something larger than default

plot(audio[6*44100:7*44100])
plt.title("This is how the 6th second of this audio looks like:")
show()
```

```python
from essentia.standard import *
w = Windowing(type = 'hann')
spectrum = Spectrum()  # FFT() would return the complex FFT, here we just want the magnitude spectrum
mfcc = MFCC()
env = Envelope()
```

```python
frame = audio[6*44100 : 7*44100]#+1024]
spec = spectrum(w(frame))

mfcc_bands, mfcc_coeffs = mfcc(spec)
env_res = env(spec)

plot(spec)
plt.title("The spectrum of a frame:")
show()

plot(mfcc_bands)
plt.title("Mel band spectral energies of a frame:")
show()

plot(mfcc_coeffs)
plt.title("First 13 MFCCs of a frame:")
show()

plot(env_res)
plt.title("Env")
show()
```
